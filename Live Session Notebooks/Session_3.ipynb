````python
# Multiple Inheritance
class Dog:
    def __init__(self):
        self.name = 'Dog'

    def bark(self):
        print("barks")
        
class Pet:
    def __init__(self):
        self.name = 'Pet'

    def runs(self):
        print("runs")

class PetDog(Pet, Dog):
     def __init__(self):
        Pet.__init__(self)
        Dog.__init__(self)
        

pet_dog = PetDog()

print(pet_dog)

pet_dog.bark()
pet_dog.runs()
print(pet_dog.name)

````
